# Visualizing Chipotle's Data

This time we are going to pull data directly from the internet.
Special thanks to: https://github.com/justmarkham for sharing the dataset and materials.

### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv).

```python
df = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep = '\t')
```

### Step 3. Assign it to a variable called chipo.

```python
chipo = df
```

### Step 4. See the first 10 entries

```python
chipo.head(10)
```

### Step 5. Create a histogram of the top 5 items bought

```python

item_counts = chipo['item_name'].value_counts().head(5)

item_counts.plot(kind='bar', color='skyblue', figsize=(8,5))
plt.title('Top 5 Most Ordered Items')
plt.xlabel('Item')
plt.ylabel('Number of Orders')
plt.show()
```

### Step 6. Create a scatterplot with the number of items orderered per order price
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
chipo['item_price'] = chipo['item_price'].str.replace('$','').astype(float)
order_price = chipo.groupby('order_id')['item_price'].sum()
order_items = chipo.groupby('order_id')['quantity'].sum()
plt.figure(figsize=(8,6))
plt.scatter(order_price, order_items, alpha=0.5)
plt.title('Number of Items Ordered per Order Price')
plt.xlabel('Order Price')
plt.ylabel('Items Ordered')
plt.show()
```

### Step 7. BONUS: Create a question and a graph to answer your own question.

```python
# 가장 많이 주문 된 음료는?
drinks = chipo[chipo['item_name'].str.contains('Drink', case=False)]

top_drinks = drinks['item_name'].value_counts().head(5)

top_drinks.plot(kind='bar', color='orange', figsize=(8,5))
plt.title('Top 5 Most Ordered Drinks')
plt.xlabel('Drink')
plt.ylabel('Number of Orders')
plt.show()
```
