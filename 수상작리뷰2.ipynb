1. 주제 및 데이터

주제 : 기업의 미래 성공 가능성을 예측

데이터는 기업의 재무 비재무 지표, 기업가치 관련 수치, 다양한 범주형, 수치형 변수로 구성되어 있으며 일부 컬럼은 결측치 비율이 상당히 높다. 전체적으로 예측값이 실제보다 높게 나오는 경향이 있어, 모델링 단계 이후 별도의 보정 과정이 필요했다.

2. 코드 리뷰

(1) EDA

변수 수가 많아 핵심 변수 중심으로 탐색을 진행했다. 기업가치 관련 변수는 실수형으로 변환해 분석 가능하도록 처리했고, 결측치가 많은 구조이기 때문에 범주형 기준으로 그룹을 나눈 뒤 그룹 평균값으로 결측치를 채우는 방식을 사용했다.

(2) 전처리

파생변수 생성에 비중을 두고 여러 조합을 시도하며 성능 변화를 확인했다. 또한 5-Fold 교차검증에서 필요 없는 컬럼을 제거하는 실험을 반복하면서 모델 복잡도를 줄이고 성능을 높였다. 전체 컬럼을 그대로 사용하는 대신 선택적으로 제거하는 전략이 적용됐다.

(3) 모델링

여러 모델을 시험했지만 최종적으로는 Random Forest 단일 모델이 가장 안정적인 성능을 보여 채택되었다. 예측 결과가 실제보다 높게 치우치는 문제가 있어, 실험적으로 산출한 보정값을 적용해 결과를 조정했다. 전 과정에서 교차검증을 바탕으로 검증이 이루어져 과적합을 잘 억제했다.

3. 차별점 및 배울 점

(1) 차별점

예측값 과대추정 문제에 대해 단순 모델 변경이 아니라 직접 보정값을 찾아 후처리했다. 복잡한 앙상블 대신 단일 모델(Random Forest)에 집중해 최적 성능을 확보한 전략도 좋았다.

(2) 배울 점

결측치 처리에서처럼, 데이터 구조에 맞는 단순한 방식을 적절히 활용해도 충분히 효과적일 수 있다. 피처 엔지니어링을 교차검증과 함께하면 과적합을 줄이고 안정적인 성능을 얻을 수 있음을 확인할 수 있다. 회귀 문제에서 예측값이 치우칠 때는 후처리 보정이 점수 개선에 크게 기여할 수 있으며, 문제에 맞는 단일 모델이 복잡한 모델 조합보다 효율적일 수 있다는 점도 배웠다.
