**1. 주제 및 데이터**

사이버 공격 유형 예측 AI 해커톤

네트워크 트래픽 로그 데이터를 기반으로 각 트래픽이 어떤 사이버 공격 유형에 해당하는지 분류

입력 변수 : 패킷 수, 바이트 수, 플로우 정보 등 트래픽 기반 변수

출력 변수 : 공격 유형 (다중 클래스 분류)

주제 : 네트워크 트래픽 데이터로부터 사이버 공격 유형을 예측하는
AI 분류 알고리즘 개발


**2. 코드 리뷰**


(1) 전처리

- 파생변수(feature engineering) 생성에 집중

- 기존 변수들을 단순히 정규화, 표준화하기보다
변수 간 조합을 통해 새로운 정보를 만들어냄

- 데이터 스케일링보다는 XGBoost가 스스로 분할 기준을 학습하도록 두는 전략

- 불필요한 전처리 단계를 최소화하여 모델이 데이터 구조를 직접 학습하도록 설계

(2) 모델링

- XGBoost 단일 모델 기반 접근

- LightGBM, CatBoost, 앙상블 등 다양한 시도를 했으나
최종적으로 XGBoost가 가장 안정적인 성능을 보임

- 과도한 멀티 모델 스태킹보다는 단일 모델 + 파생변수 조합으로 점수 향상

- 하이퍼파라미터 튜닝보다 피처 구성 자체가 성능에 더 큰 영향을 미침을 확인

**3. 차별점 및 배울 점**

- 모델 구조보다 피처 엔지니어링이 성능을 좌우할 수 있음을 보여줌

- 복잡한 앙상블이나 딥러닝보다 전통적인 트리 기반 모델 + 잘 설계된 변수가 효과적일 수 있음

- 여러 모델을 무작정 쌓기보다는 가장 잘 작동하는 하나의 모델을 깊게 파는 전략이 유효했음
