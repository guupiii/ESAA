**1. 주제 및 데이터**

신용카드 고객 세그먼트 분류를 주제로 한 데이콘 대회

데이터는 약 800여 개의 변수, 고객의 카드 이용 패턴을 바탕으로 A~D 네 개 세그먼트 중 하나로 예측하는 문제이다.

평가 지표는 Macro F1-score이다. 클래스 간 불균형이 뚜렷하다

**2. 코드 리뷰 **

(1) EDA
- 변수의 수가 많아 주요 변수 중심으로 탐색을 진행
- EDA를 통해 클래스 불균형이 심하다는 점을 확인하고, 특정 세그먼트(A/B, C/D)의 분포 차이에 주목

(2) 전처리
- 클래스 불균형 문제를 해결하기 위해 SMOTE 오버샘플링 및 class_weight 조정을 수행
- 세그먼트별 특성을 고려하여 A/B 세그먼트만 따로 학습하는 VIP 모델을 구성.
-> 클래스 간 혼동을 줄이고 주요 그룹의 예측 성능을 향상시킴

(3) 모델링
- 다양한 모델과 앙상블 기법을 실험했으나, 최종적으로 CatBoost 단일 모델이 가장 높은 성능을 보여 선택됨
- 모델 파라미터 최적화 외에도, C/D 클래스 구간의 예측 성능을 개선하기 위해 일부 세그먼트를 제거하고 재학습
- 훈련 및 검증 과정에서는 GPU 환경(A100, L4)을 활용하여 학습 효율을 높임

**3. 차별점 및 배울 점**

(1) 차별점
- 모든 클래스를 한 번에 학습하는 대신 특정 그룹(A/B)을 독립적으로 학습하여 성능을 개선하엿다.
- SMOTE와 class_weight를 병행하였다.
- 복잡한 앙상블보다 CatBoost 단일 모델이 더 높은 성능을 보였다.

(2) 배울 점
- 클래스별 특성에 따른 모델 분리가 실제 성능 향상으로 이어질 수 있음을 확인할 수 있었다.
- 불균형 데이터의 경우 단순한 샘플링 외에도 class_weight 조정이 효과적일 수 있다.
- 모델 단순화와 핵심 피처 집중이 성능 향상에 중요하다.
